```python
# Import libraries 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Boundary and initial conditions
#Temperatures
T_Ar = 293 #K
T_LN2 = np.linspace(80,100,20) #K

# Thermal/material properties
k_SS = 14.4 #W/mK
k_Cu = 386 #W/mK
k_Ar = 0.016 #W/mK 

# Dimensions
t_SS = 0.0012446 #m (0.049")
r_SS = 0.00635/2 #m (1/8")
L_SS = 0.1016 #m (4")
r_Cu = 0.0381 #m (1.5")
L_Cu = 0.0254 #m (1")
W_Cu = 2*r_Cu #m (3")
```

```python
# Caclulate Areas
A_SS = 2*np.pi*r_SS*L_SS #m^2
A_Cu_Cyl = 2*np.pi*(r_Cu**2) + 2*np.pi*r_Cu*L_Cu #m^2
A_Cu_Blk = 2*(W_Cu**2) + 4*(W_Cu*L_Cu) #m^2
A_Ar = 2*np.pi*(0.0254/2)*0.3254 #m^2

# Calculate Thermal Resistances
R_SS = L_SS/(k_SS*A_SS) #K/W
R_Cu_Cyl = L_Cu/(k_Cu*A_Cu_Cyl) #K/W
R_Cu_Blk = L_Cu/(k_Cu*A_Cu_Blk) #K/W
R_Ar = 1/(2*np.pi*(0.0254/2)*k_Ar)
R_tot_Cyl = R_SS+R_Cu_Cyl+R_Ar
R_tot_Blk = R_SS+R_Cu_Blk+R_Ar
```

```python
# Calculate total heat flux
Q_Cyl = (T_Ar - T_LN2)/(R_tot_Cyl)
Q_Blk = (T_Ar - T_LN2)/(R_tot_Blk)

# Plot heat load as a function of the LN2 temperature
QvT = plt.figure(figsize=(10,10))
QvT = plt.title("Heat flux from LN2 temperature")
QvT = plt.xlabel("LN2 temperature (K)")
QvT = plt.ylabel("Heat flux (W)")
QvT = plt.scatter(T_LN2,Q_Cyl,c="blue",label="Cylinder",s=10)
QvT = plt.scatter(T_LN2,Q_Blk,c="red",label="Block",s=10)
QvT = plt.plot(T_LN2,Q_Cyl,color='blue')
QvT = plt.plot(T_LN2,Q_Blk,color='red')
QvT = plt.legend(loc='best')
QvT = plt.show()
```

```python
# Calculate the thermal resistances from both sides
R_cold_Blk = R_SS + R_Cu_Blk #K/W
R_cold_Cyl = R_SS + R_Cu_Cyl #K/W
R_hot = R_Ar #K/W

# Calculate the temperature of the copper block for the block and cylinder
T_Cu_cold_Blk = T_LN2 + Q_Blk*(R_Cu_Blk+R_SS)
T_Cu_hot_Blk = T_Ar - Q_Blk*R_hot
T_Cu_cold_Cyl = T_LN2 + Q_Cyl*(R_Cu_Cyl+R_SS)
T_Cu_hot_Cyl = T_Ar - Q_Cyl*R_hot

# Plot the final temperatures for each configuration as a function of LN2 temperature
TvT, (Blk, Cyl) = plt.subplots(1,2,figsize=(20,10))
Blk.set_title("Temperature of Rectangular Cu Block")
Blk.set(xlabel='LN2 temperature (K)',ylabel='Cu temperature (K)')
Blk.plot(T_LN2,T_Cu_cold_Blk,color='blue',label="Cold Side")
Blk.scatter(T_LN2,T_Cu_hot_Blk,color='red',label="Hot Side")
Blk.legend()
Cyl.set_title("Temperature of Cylindrical Cu Block")
Cyl.set(xlabel='LN2 temperature (K)',ylabel='Cu temperature (K)')
Cyl.plot(T_LN2,T_Cu_cold_Cyl,color='blue',label="Cold Side")
Cyl.scatter(T_LN2,T_Cu_hot_Cyl,color='red',label="Hot Side")
Cyl.legend()
```

```python
#=========================IF WE ASSUME THERMAL CONVECTION DOMINATES IN GAr=========================#
# Calculate convective coefficient 
# Nu = hD/k => h = kNu/D
# Nu = 3.66 + ((0.065)(D/L)RePr))/(1+(0.04)((D/L)RePr)^(2/3))
# Re = 4 m_dot /pi D mu
# Pr = c_p mu / k

D = 0.0254 #m
L = 0.3254 #m
c_p = 520 #J/kg/K
mu = 2.42E-5 #kg/m/s
m_dot = 8.32E-5 #kg/s (slpm)

Pr = (c_p*mu)/k_Ar 
Re = (4*m_dot)/(np.pi*D*mu)
Nu = 3.66 + (0.065*(D/L)*Re*Pr)/(1+0.04*(((D/L)*Re*Pr)**(2/3)))
h_Ar = (Nu*k_Ar)/D

Resistance = 1/(h_Ar*A_Ar)
```

```python
Q_dot = (T_Ar-T_LN2)/Resistance

# Plot the heat flux as a function of LN2 temperature 
QvT = plt.figure(figsize=(10,10))
QvT = plt.title("Heat flux as a function of LN2 temperature")
QvT = plt.xlabel("LN2 temperature (K)")
QvT = plt.ylabel("Heat flux (W)")
QvT = plt.scatter(T_LN2,Q_dot)
QvT = plt.plot(T_LN2,Q_dot)
QvT = plt.show()
```

```python
# Calculate the final temperature
T_cold = T_LN2 + Q_dot*(R_Cu_Cyl+R_SS)
T_hot = T_Ar - Q_dot*Resistance

# Plot the temperatures from the hot and cold sides
TvT = plt.figure(figsize=(10,10))
TvT = plt.title("Cu temperature as a function of LN2 temeperature")
TvT = plt.xlabel("LN2 temperature (K)")
TvT = plt.ylabel("Cu temperature (K)")
TvT = plt.plot(T_LN2,T_cold,c='blue',label='Cold Side')
TvT = plt.scatter(T_LN2,T_hot,c='red',label='Hot Side')
TvT = plt.legend(loc='best')
TvT = plt.show()
```
